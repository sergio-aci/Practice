# Intro
`matplotlib` is the plotting library of choice for data science in python. There are other fancy libraries for specific use cases and for specifc styles. But, you'll have to work really hard to find a use case matplotlib can't cover. 

`matplotlib` is a python library used to create 2D graphs and plots by using python scripts. It has a module named `pyplot` which makes things easy for plotting by providing feature to control line styles, font properties, formatting axes etc. It supports a very wide variety of graphs and plots namely - histogram, bar charts, power spectra, error charts etc. It is used along with NumPy to provide an environment that is an effective open source alternative for MatLab.

Worth noticing that there are also other well known libraries:
- [Seaborn](http://seaborn.pydata.org/index.html#Seaborn): more aesthetically pleasing and looks more modern. Built on top of matplotlib.
- [Bokeh](https://docs.bokeh.org/en/latest/index.html): Interactive, web-ready plots which can be output as JSON objects, HTML documents or interactive web apps.
- [Plotly](https://plotly.com/): also interactive plots with some charts you won't find in most libraries (contours plots, dendograms, 3D charts)
- [geoplotlib](https://github.com/andrea-cuttone/geoplotlib): for creating maps and plotting geographical data.

# Plotting Lines
as you've seen in the previous example to plot a line you use the `plot` function. Let's see another example

```python
x = np.array([1, 10, 100, 200])
y = x**2

plt.plot(y)
plt.title('Too much power');
```

What if we want to plot `x` against `y`?
You can call the `plot` function with or without the `x` parameter. 

```python
plt.plot(x, y)
plt.title('Too much power');
```

**Note** if you like you can change the x axis, y axis or both to a logarithmic scale using `semilogx, semilogy, loglog`

```python
x = np.array([1, 10, 100, 200])
y = x**2

f = plt.figure()
plt.semilogy(x, y)
plt.title('Too much power')
plt.show()
```

## Colors and Styles
you can specify the colors and marker shapes for line plots either with keyword arguments or with a formatted string as positional argument

We can easier call `marker`, `color` etc. explicitly, or use format `fmt` parameter in the following format `'[marker][line][color]'`

```python
x = np.arange(10)

plt.plot(x, x, marker='+', color='red', label='keyword arguments')
plt.plot(x, x+1, '.-b', label='formatted string')
plt.legend();
```

We can also have no line, or no markers etc.:

```python
x = np.arange(10)

plt.plot(x, x, color='magenta', label='keyword arguments')
plt.plot(x, x+1, '.k', label='formatted string')
plt.legend();
```

# Scattered
lines are great but lose their appeal after a while. First let's see a motivational example describing the location and quality of food on a 2D plane

```python
import pandas as pd

# generating some random data
n = 50

# coordinates
x = np.random.normal(0, 0.1, n) + np.random.randint(-1, 1, n)
y = np.random.normal(0, 0.13, n) + np.random.randint(-1, 1, n)

# food features
tasty = np.random.randint(0, 5, n) * 100
healthy = np.random.poisson(1, n) * 100

looking_for_food = pd.DataFrame({'x': x, 'y':y, 'tasty': tasty, 'healthy': healthy})
looking_for_food.head()
```

All these numbers are making me dizzy. Where's the food? Let's find out using a scatter plot. A scatter plot is a 2D plot for showing how points are scattered on a 2D plane. Run the next example to see it in action.

```python
looking_for_food['x'].head(5) # is a Series
```

```python
#f = plt.figure()

plt.scatter(looking_for_food['x'], looking_for_food['y'])
plt.show()
```

OK. This is quite informative. We now have a sense of where the food is at and how spread out is it. But... is it tasty? How can we use a 2D plot to show us? It's not hard. Let's add some color with the `c` keyword to scatter.

```python
f = plt.figure()

plt.scatter(looking_for_food['x'], looking_for_food['y'], c=looking_for_food['tasty'])
plt.show()
```

This is nice and cheery, but what do the colors mean? So much like labels and the `legend` command, colors are not very useful if we don't show a colorbar using the `plt.colorbar` function.

```python
looking_for_food
```

```python
plt.scatter(looking_for_food['x'], looking_for_food['y'], c=looking_for_food['tasty'])
plt.colorbar()
plt.show()
```

That's really nice, but what about our health? I want to see that too. Let's do the same trick we did for taste but using sizes with the `s` keyword to scatter. 

```python
plt.scatter(looking_for_food['x'], 
            looking_for_food['y'], 
            c=looking_for_food['tasty'], 
            s=looking_for_food['healthy'])
plt.colorbar()
plt.show()
```

In order to be able to see the colors that are one on top of the other better, let's add opacity with `alpha`:

```python
f = plt.figure()

plt.scatter(looking_for_food['x'], 
            looking_for_food['y'], 
            c=looking_for_food['tasty'], 
            s=looking_for_food['healthy'],
            alpha=0.4,
           cmap="jet_r")
plt.colorbar()
plt.show()
```

```python
f = plt.figure()

plt.scatter(looking_for_food['x'], 
            looking_for_food['y'], 
            c=looking_for_food['tasty'], 
            s=looking_for_food['healthy'],
            alpha=0.4,
           cmap="plasma_r")
plt.colorbar()
plt.show()
```

## Multiple plots
We can also make multiple plots (figures!) one after the other in a cell. Check it out

```python
plt.figure()
plt.plot(looking_for_food['healthy'].sort_values(), looking_for_food['tasty'].sort_values())
plt.title('Taste Vs. Health')
plt.show()

print('This separates between the figures')

plt.figure()
plt.scatter(looking_for_food['x'], looking_for_food['y'])
plt.title('Where\'s the food')

plt.show()
```

That's nice. But the first plot is a bit unclear. Which axis refers to taste and which to health? We can label plot axes using the `plt.xlabel` and `plt.ylabel` functions.

```python
plt.figure()
plt.plot(looking_for_food['healthy'].sort_values(), looking_for_food['tasty'].sort_values())
plt.title('Taste Vs. Health')
plt.xlabel('Health')
plt.ylabel('Taste')

plt.figure()
plt.scatter(looking_for_food['x'], looking_for_food['y'])
plt.title('Where\'s the food')

plt.show()
```

#### Till now we have been working with `figure` and `axes` implicit (Matlab style), which is fine when you have relatively simple code.   For more complicated code, we will need to work Object Oriented style (explicit `figure`, but especially explicit `axes`)

```python
fig = plt.figure() # 1. create a figure
axes1 = fig.add_axes((0, 0.8, 1, 1)) #2. create subplots [left, bottom, width, height] fraction
axes2 = fig.add_axes((0.7, 0.8, 0.2, 0.3))

axes1.plot([2, 10, 100, 1000, 7]) #3. plot smth
axes2.plot([1, 5, 3, 6, 7])
plt.show()
```

## Object oriented vs. non Object Oriented way of working

#### Non Object Oriented way
- Use `pyplot` calls directly as we have seen till now
- More concise
- Current `figure` and `axes` is implicit, saved in memory, calls done on `pyplot` are forwarded to them
- Better for very simple graphs, especially with only 1 `axes` \ `subplots`
- Slightly more confusing to understand relationships

#### Object oriented way
- Use a call from the relevant level, like `figure`, and mostly `axes`
- Not as concise
- Easier to understand relationships and what object we are working on
- Better for multiple `axes` \ `subplots` and for fine tuning our graphs

You will see both paradigms, therefore important to understand both.  To understand how `matplotlib` works, need to understand OO way and relationships between main entities.

## Subjugate plots - subplots
instead of using multiple figures we can use subplots for which we can control alignment. Calling `plt.subplots` returns a tuple of `figure`, numpy array of `axes`.  A figure might have multiple `axes` (individual graphs), we plot on the specific `axes`

```python
plt.subplots(2, 2)
```

```python
f, ax = plt.subplots(2, 2)
ax[0,1] # top right subplot
ax[0,0] # top left subplot
# ...
```

```python
f, ax = plt.subplots(2, 1) # very important to remember
print('Axis array len: ', len(ax), 'type: ', type(ax))

ax[0].plot(looking_for_food['healthy'].sort_values(), \
           looking_for_food['tasty'].sort_values())

# plt.show()
print("This doesn't separate anything")

ax[1].scatter(looking_for_food['x'], looking_for_food['y'])

plt.show()
```

Almost any plot decoration which is called by `plt.<command>` has a counterpart `ax.<command>` or `<ax>.set_<command>` for calling it on subplot `ax`. 

<div class="alert alert-info">
<b>Note:</b> the ax object returns by `plt.subplots` is an array accessed in the same way you would access a numpy array.
</div>

The graph above is not bad. It would be better if each subplot would have a title and labeled axes, and be side by side.
Let's:
1. Set title for each subplot
2. Set labels for the axes of each subplot
3. Set the subplots side by side

```python
f, ax = plt.subplots(1, 2)

ax[0].plot(looking_for_food['healthy'].sort_values(), looking_for_food['tasty'].sort_values())
ax[1].scatter(looking_for_food['x'], looking_for_food['y'])

ax[0].set_title('Health vs Tasty')
ax[0].set_xlabel('health')
ax[0].set_ylabel('taste')

ax[1].set_title('x vs y') 
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

plt.show()
```

Instead of writing 
1. `ax.set_<command1>(<arg1>)`
2. `ax.set_<command2>(<arg2>)`
3. ...

We can use `ax.set(<command1>=<args1>, <command2>=<args2>,...)`

Let's use this format to plot the previous subplots with titles & axis labels

<div class="alert alert-info">
<b>Note:</b> If you are tired of the boring old grid, you can create custom subplot alignment using `plt.GridSpec`
</div>

```python
f, ax = plt.subplots(1, 2)

ax[0].plot(looking_for_food['healthy'].sort_values(), looking_for_food['tasty'].sort_values())
ax[1].scatter(looking_for_food['x'], looking_for_food['y'])

#ax[0].set_title('Health vs Tasty')
#ax[0].set_xlabel('health')
#ax[0].set_ylabel('taste')
ax[0].set(title='Health vs Tasty', xlabel='health', ylabel="taste")

#ax[1].set_title('x vs y') 
#ax[1].set_xlabel('x')
#ax[1].set_ylabel('y')
ax[1].set(title='x vs y', xlabel='x', ylabel="y")

plt.show()
```

# Histograms
Histograms are an awesome tool to explore distributions of samples in sample space. Let's start with some motivation. 

Let's work with a larger sample size.

```python
import pandas as pd

# generating some random data
n = 5000

# coordinates
x = np.random.normal(0, 0.1, n) + np.random.randint(-1, 1, n)
y = np.random.normal(0, 0.13, n) + np.random.randint(-1, 1, n)

# food features
tasty = np.random.randint(0, 5, n) * 100
healthy = np.random.poisson(1, n) * 100

looking_for_food = pd.DataFrame({'x': x, 'y':y, 'tasty': tasty, 'healthy': healthy})
looking_for_food.head()
```

```python
import pandas as pd

# generating some random data
n = 5000

# coordinates
rng = np.random.default_rng()
x = rng.normal(0, 0.1, n) + rng.integers(-1, 1, n)
y = rng.normal(0, 0.13, n) + rng.integers(-1, 1, n)

# food features
tasty = rng.integers(0, 5, n) * 100
healthy = rng.poisson(1, n) * 100

looking_for_food = pd.DataFrame({'x': x, 'y':y, 'tasty': tasty, 'healthy': healthy})
looking_for_food.head()
```

Let's try to present the data with a scatter plot the same way we did before

```python

plt.scatter(looking_for_food['x'],
            looking_for_food['y'],
            c=looking_for_food['tasty'],
            s=looking_for_food['healthy'],
            alpha=0.3)
plt.show()
```

That's not very informative, is it? It just looks like a set of blobs...

Here is exactly where histograms come into play. Histograms assignes each data point into a bin and presents us with size of each bin. 

Let's:
1. Use `plt.hist` to plot the x values with 2, 13 and 100 bins
2. Do the same but for y+x

<div class="alert alert-info">
    <b>Note:</b> Use subplots and a loop to keep the plots and code tight
</div>

```python
list_bins_sizes = [2, 13, 20, 100]

fig1, ax = plt.subplots(1, len(list_bins_sizes), figsize=(15,5))
for i, num_bins in enumerate(list_bins_sizes):
    ax[i].hist(x, bins=num_bins)
    ax[i].set(title=str(num_bins)+' bins', xlabel='x')

fig1.tight_layout()

fig2, ax2 = plt.subplots(1, len(list_bins_sizes), figsize=(15,5))
for i,num_bins in enumerate(list_bins_sizes):
    ax2[i].hist(x+y, bins=num_bins)
    ax2[i].set(title=str(num_bins)+' bins', xlabel='x+y')

fig2.tight_layout()
plt.show()
```

I hope we are in agreement that histograms are great. And the greatness doesn't stop here. `plt.hist` actually returns the number of elements per bin and the bin edges as return parameters.

## Adding some words
while a picture is worth a thousand words you still may want to add some words to your plots. You can achieve this using the `plt.text` command. 

Let's:

Plot one of the histogtrams again and add some text to highest bin in it

```python
plt.hist(x, bins=100)
plt.text(-0.75,20,'SOME TEXT HERE')
plt.show()
```

Sometimes we want to show to which point exactly our text refers to. To achieve this we can use the `plt.annotate` function. This function puts some text with an arrow pointing to some point in the plot.

Let's:

Repeat the previous exercise but now with annotation. You'll need to set the following parameters
1. `xy` for where to point
2. `xytext` for where to put the text
3. `s` the text
4. `arrowprops` a dictionary with arrow properties (e.g., `arrowprops={'arrowstyle':"->"}`)

```python
plt.hist(x, bins=100)
plt.annotate('blabla', xy=(-1,180), xytext=(-0.6,150), arrowprops={'arrowstyle':"->"})
plt.show()
```
